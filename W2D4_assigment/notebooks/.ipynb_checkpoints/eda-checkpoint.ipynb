Exploratory Data Analysis on analytics_table.

```python
from pathlib import Path
import sys


ROOT = Path().resolve().parent
if str(ROOT / 'src') not in sys.path:
    sys.path.insert(0, str(ROOT / 'src'))

import pandas as pd
from data_workflow.viz import bar_sorted, time_line, histogram_chart, save_fig
from data_workflow.utils import bootstrap_diff_means


DATA = ROOT / 'data/processed/analytics_table.parquet'
FIGS = ROOT / 'reports/figures'
FIGS.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_parquet(DATA)
df.head()
```

## Revenue by country

```python
rev = df.groupby('country', dropna=False)['amount'].sum().reset_index()
fig = bar_sorted(rev, 'country', 'amount', 'Revenue by Country')
save_fig(fig, FIGS / 'revenue_by_country.png')
fig
```

## Monthly revenue trend

```python
trend = df.groupby('created_at_month')['amount'].sum().reset_index().sort_values('created_at_month')
fig = time_line(trend, 'created_at_month', 'amount', title='Monthly Revenue Trend')
save_fig(fig, FIGS / 'monthly_trend.png')
fig
```

## Amount distribution

```python
fig = histogram_chart(df, 'amount_wins', nbins=30, title='Distribution of Order Amounts')
save_fig(fig, FIGS / 'amount_hist.png')
fig
```

## Bootstrap comparison example

```python
a = df.loc[df['country']=='SA','amount']
b = df.loc[df['country']=='AE','amount']
res = bootstrap_diff_means(a,b)
res
```
